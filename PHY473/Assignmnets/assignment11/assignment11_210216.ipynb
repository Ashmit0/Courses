```python
# define RK4 method ! 

def RK4( x0 , xn , dx , y0 , f , e ): 
    x = np.arange( x0 , xn , dx )
    y = np.zeros( (x.size , y0.size ))
    y[0] = y0 
    for i in range( 1 , x.size ): 
        rk1 = dx*f( y[i-1] , x[i-1] ,e )
        rk2 = dx*f( y[i-1] + rk1/2 , x[i-1] + dx/2 , e )
        rk3 = dx*f( y[i-1] + rk2/2 , x[i-1] + dx/2 , e )
        rk4 = dx*f( y[i-1] + rk3 , x[i-1] + dx , e )
        y[i] = y[i-1] + ( rk1 + 2*rk2 + 2*rk3 + rk4 )/6 
    return y 
```

```python
# defineie f function : 
# y0 -> psi , v 

def f( y , x , e ): 
    dp = y[1]
    dv = -(2*m/hbar**2)*(e - V0*(x**2)/(a**2) )*y[0]
    return np.array([dp,dv])

y0 = np.array([0 , 1 ])

yend = []
elist = np.arange( 100*eV , 200*eV , 1*eV )
for e in elist : 
    y = RK4( -10*a , 10*a , 0.01*a , y0 , f , e )
    yend.append( y[-1 , 0 ]**2  )
    # if( y[ -1 , 0 ] < 0.00001 ) : 
    #     print( e/eV )
```

```python
# define the swap function 
def swap( A , i ): 
    m = A.shape[0]
    if( i >= m - 1 ): 
        return A 
    else : 
        if( A[ i ,i ] == 0 ): 
            j = i 
            while A[ j , i ] == 0 : 
                j += 1 
            A[[i,j]] = A[[j,i]]
        return A 
```

```python
# define the function that makes the matrix upper triangular 
def upper( A  ):
    m = A.shape[0]
    for i in range( m ): 
        A = swap( A, i )
        A[i,:]/=A[i,i]
        j = i + 1 
        while j < m : 
            A[j,:] -= A[i,:]*A[j,i]
            j += 1 
    return A
```

```python
def back_solve( A ): 
    m = A.shape[0]
    result = np.zeros(m)
    for i in range( m - 1, -1 , -1 ):
        result[i] = A[ i , m ]
        for j in range(i + 1 , m ): 
            result[i] -= A[i,j]*result[j]
    return result 
```

```python
A = np.array([
    [0,0,2,1,2],
    [0,1,0,2,-1],
    [1,2,0,-2,0],
    [0,0,0,-1,1],
    [0,1,-1,1,-1]
] , dtype = np.float32 )
print( A , A.shape )
```

```python
V = np.array([1 , 1, -4, -2 , -1 ]).reshape((5,1))
print( V , V.shape )
```

```python
def gaus( A , V ): 
    A = np.concatenate( (A , V) , axis = 1 )
    A = upper( A )
    result = back_solve( A  ) 
    result = result.reshape((result.size , 1 ))
    return result 

result = gaus( A , V )
print( f'result = \n {result}' )
```

```python
def lu_decomp( A ): 
    n = A.shape[0]
    L  = np.zeros( ( n , n ))
    U = np.zeros( (n , n ))
    U[0,:] = A[0,:]
    for i in range( n ): 
        L[ i , i ] = 1 
    for i in range( 1 , n ): 
        for j in range(0 , n ): 
            if( i > j ):
                L[i,j] = A[i,j] 
                for k in range( j ): 
                    L[i,j] -= L[i,k]*U[k,j]
                L[i,j] /= U[j,j]
            else: 
                U[i,j] = A[i,j]
                for k in range( i ): 
                    U[i,j] -= L[i,k]*U[k,j]
    return L , U 
```

```python
A = np.array([
    [4,-3,6],
    [8,-3,10],
    [-4,12,-10]
])
L , U = lu_decomp( A )

print( f'L = {L} \n and \n U = {U}')
```

```python
def forward_solve( A , V ): 
    m = A.shape[0]
    result = np.zeros( m )
    for i in range( m ): 
        result[i] = V[i]
        if i > 0 : 
            for k in range( i ): 
                result[i] -= result[k]*A[i , k ]
        result[i] /= A[i,i]
    return result 
```

```python
def back_solve( A , V ): 
    m = A.shape[0]
    result = np.zeros(m)
    for i in range( m - 1, -1 , -1 ):
        result[i] = V[i] 
        for j in range(i + 1 , m ): 
            result[i] -= A[i,j]*result[j]
        result[i] /= A[i,i]
    return result 
```

```python
def lu_solve( A , V ): 
    L , U = lu_decomp( A )
    n = V.shape[1]
    result = np.zeros(  V.shape )
    for i in range( n ): 
        result1 = forward_solve( L , V[:,i].reshape( - 1, 1 ) )
        result2 = back_solve( U , result1.reshape( -1 , 1 ) )
        result[ : , i ] = result2 
    return result 
    
V = np.array([
    [1,0],
    [0,1],
    [0,0]
])
result  = lu_solve( A , V )
print( f'result = \n {result}' )
```
