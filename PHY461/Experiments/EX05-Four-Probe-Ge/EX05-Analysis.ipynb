```python
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd 

from scipy.stats import linregress
```

```python
# set default params for matplotlib : 
plt.rcParams['font.size'] = 14
plt.rcParams['lines.linewidth'] = 2
plt.rcParams["figure.figsize"] = (12 , 8 )
```

```python
data = pd.read_excel('data.xlsx' , sheet_name=0)
```

```python
x = data['1000/T(1/K)']
y = np.log( (data['rho Heating'].to_numpy() + data['rho Cooling'].to_numpy() )/2)

idx = np.where( x <= 3.105  )[0][0]
plt.scatter( x[:idx] , y[:idx] , marker='*', linewidth = 2.5 ,label='Extrinsic Region')
plt.xticks( x , rotation = -90  )

x , y = x[idx:] , y[idx:]

slope , intercept   = linregress( x , y )[:2]
eg = slope * 0.172 

plt.scatter( x , y , marker='s' , color = 'g' , linewidth = 2.5 ,label='Intrinsic Region')
plt.plot( x , x*slope + intercept , 'r' , label = 'Line of Best Fit')
plt.xlabel(r'$\frac{1000}{T}$ ($\frac{1}{K}$)')
plt.ylabel(r'$\ln \rho(T)$')
plt.title('log Resistivity of Ge Crystal vs Inverse Temperature at 5mA')
plt.legend()
plt.grid()
plt.text( 2.3 , -2 , r"$E_g$ = {:.4f} $eV$".format(eg) , fontsize=16, color='black')
plt.show()
```
