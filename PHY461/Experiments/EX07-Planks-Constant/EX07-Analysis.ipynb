```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 

from scipy.stats import linregress

# set default params for matplotlib : 
plt.rcParams['font.size'] = 18
plt.rcParams['lines.linewidth'] = 2
plt.rcParams["figure.figsize"] = (12,8)
```

```python
data = pd.read_excel('data.xlsx' , sheet_name=1)
data
```

```python
y , x = data['Mean(mV)'] , data['Wavelength(nm): ']
x = 3000/x


result = linregress( x , y )
slope , intersept = result[:2]
h = 1.60217646*slope/100
```

```python
plt.plot( x , y, '-o' , label = 'Measured Data Points')
plt.plot( x , x*slope + intersept , 'r--' , label = 'Line of Best Fit')
plt.xlabel(r'Frecquency of Monochromatic Light $\nu$ ($10^{14} Hz$)')
plt.ylabel(r'Mean Voltage Measured $U$ ($mV$)')
plt.xticks(x)
plt.yticks(list(range(0,41,10)) + list(y))
plt.grid()
plt.legend()
plt.text(
 4.67 , 36 , 
 r"$h = {:.4f} \times 10^{{-34}}$".format(h) , 
 fontsize=20, color='black'
)
plt.text(
 4.67 , 32, 
 r'$h = e \times \frac{\mathrm{d}U}{\mathrm{d}v} = e \times \text{slope}$' , 
 fontsize=20, color='black'
)
plt.title('Plot for Table 1 Readings:'); 
```

```python
result.stderr*1.60217646/100
```
