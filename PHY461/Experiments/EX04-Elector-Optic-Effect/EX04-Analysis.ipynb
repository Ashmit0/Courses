```python
# import necessary libraries 
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
from scipy.stats import linregress
```

```python
# set default params for matplotlib : 
plt.rcParams['font.size'] = 14
plt.rcParams['lines.linewidth'] = 2
plt.rcParams["figure.figsize"] = (12,7)
```

```python
part1_data = pd.read_excel('data.xlsx' , sheet_name=0)
part2_data = pd.read_excel('data.xlsx' , sheet_name=1)
```

### Analysis for part 1 

```python
voltage = part1_data['Source Voltage (kV)'].to_numpy()
intensity = part1_data['Intensity'].to_numpy()

imax = np.max( intensity )
index  =  np.where(intensity == imax )[0]
vmax = np.sum(voltage[ index ])/index.size 

imin = np.min(intensity[50:])
idx = np.where( intensity[50:]== imin )[0]
idx = idx + 50 
vmin = np.sum( voltage[idx] )/idx.size 


plt.plot( voltage , intensity )
plt.scatter(voltage  , intensity )
plt.scatter( vmax , imax , color = 'r' , marker = '|' , s = 300 ,  label = rf'$V_{{\lambda/2}} = {vmax:.2f}$')
plt.scatter( vmin , imin , color = 'r' , marker = '*' , s = 300 ,  label = rf'$V_{{\lambda}}$  = {vmin:.2f}')
plt.grid()
plt.xlabel(r'Source Voltage to Sample $V_s$ (KV)')
plt.ylabel(r'Intensity Readings from the Photodiode $I$')
plt.legend()
plt.title('Source Voltage vs Intensity for Part 1')
plt.show()

print( vmin , vmax )
```

### Analysis for Part 2 

```python
voltage = part2_data['Power Supply Voltage (V)']
amplitude = part2_data['Square Wave Amplitude (mV)']



plt.plot( voltage , amplitude )
plt.scatter(voltage  , amplitude )
plt.grid()
plt.xlabel(r'Source Voltage to Sample $V_s$ (V)')
plt.ylabel(r'Amplitude Readings from the CRO ($mV$)')
plt.title('Source Voltage vs Square Wave Amplitude for Part 2')
plt.show()
```
